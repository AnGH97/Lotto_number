```python
#num list와 count list를 리스트로 합치기
def mcnt_list(num_list, count_list):
    r = list()
    for i in range(0, len(num_list)):
       r.append([num_list[i], count_list[i]])
    return r
```

```python
#리스트의 원소들을 float에서 int로 변경
def ftoi(cn_list):
    ncnt_list = np.array(cn_list)
    ncnt = np.nan_to_num(ncnt_list, nan = -1)
    rcnt_list = ncnt.tolist()
    for i in range(0, len(rcnt_list)):
        for j in range(0, 2):
            rcnt_list[i][j] = int(rcnt_list[i][j])
    return rcnt_list
```

```python
#제일 많이 나온 숫자 검색
def maximum_num(cnt_list):        
    max_t = 0
    num_t = 0
    for i in range(0, len(cnt_list)):
        for j in range(1, 2):
            if(max_t < cnt_list[i][j]):
                max_t = cnt_list[i][j]
                num_t = cnt_list[i][0]
    return [num_t, max_t]
```

```python
#axis = 0 행
#axis = 1 열
#중간 숫자 뽑기
def mid_num(cnt_list):
    rcnt_list = list()
    mnum = 0
    scnt_list = sorted(cnt_list, key = lambda x:-x[1])
    for i in range(0, 45):
        for j in range(1, 2):
            if scnt_list[i][j] != -1:
                rcnt_list.append(scnt_list[i])
            else:
                continue

    if len(rcnt_list) % 2 == 0:
        index = len(rcnt_list)//2
        mnum = rcnt_list[index]
    elif len(rcnt_list) % 2 != 0:
        index = len(rcnt_list)//2 +1
        mnum = rcnt_list[index]
    return mnum
    
```

```python
#단순 랜덤으로 7개 숫자 고르기

def random_choice_num():
    num_list = list()
    while(True):
        if len(num_list) == 6:
            num_list.sort()         
            break      
        elif len(num_list) == 0:
            num = random.randrange(1, 45)
            num_list.insert(len(num_list), num)
        else:    
            num = random.randrange(1, 45)
            num_list.insert(len(num_list), num)
            for j in range(0, len(num_list)-1):
                if num == num_list[j]:
                    num_list.pop()
                    break 
    return num_list
```

```python
def link_data(ldata, on):
    link_list = list()
    for i in range(0, len(ldata)):
        for j in range(0, 1):
            if ldata[i][j] == on:
                link_list.insert(len(link_list), ldata[i])
    next_on = random.sample(link_list, 1)
    return next_on

```
