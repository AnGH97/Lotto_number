```python
import io
import pandas as pd
import numpy as np
import math
import random
```

```python
data = pd.read_csv(io.BytesIO(myfile['num720.csv']))
data
```

```python
sdata = data.drop(['round'], axis = 1)
print(sdata)
```

```python
jo = data['jo'].value_counts()
num1 = data['num1'].value_counts() 
num2 = data['num2'].value_counts() 
num3 = data['num3'].value_counts() 
num4 = data['num4'].value_counts() 
num5 = data['num5'].value_counts() 
num6 = data['num6'].value_counts() 
print(num1)
```

```python
#정렬 완료(jo)
jdf = pd.DataFrame(data = jo.index, columns = ['jo'])
cdfj = pd.DataFrame(data = jo.values, columns=['jcount'])

df = pd.merge(jdf, cdfj, left_index=True, right_index=True)
df = df.sort_values(by = 'jo')
print(df)
```
